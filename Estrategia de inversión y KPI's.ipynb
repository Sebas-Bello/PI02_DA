# **<font color="green">Estrategia de Inversión: Dollar-Cost Averaging (DCA)</font>**


La estrategia de **Dollar-Cost Averaging (DCA)**, también conocida como "Promediación en Costos", es una táctica popular para invertir en criptomonedas y otros activos. Esta estrategia se basa en adquirir criptomonedas de manera continua y disciplinada a lo largo del tiempo, independientemente de los cambios en los precios del mercado.

**Funcionamiento de la Estrategia:**

1. **Frecuencia de Compra:** En lugar de invertir una gran cantidad de dinero en una sola transacción, con DCA decides invertir una **cantidad fija en intervalos regulares**, como semanal o mensualmente.

2. **Compras en Diferentes Niveles de Precio:** Con cada compra programada, adquieres **la misma cantidad de criptomoneda**, sin importar su precio en ese momento. Esto resulta en la adquisición de más unidades cuando el precio es bajo y menos cuando es alto.

3. **Promediación de Costos:** A lo largo del tiempo, al realizar compras en diferentes niveles de precio, tu **costo promedio por unidad disminuirá**, ya que estarás comprando más unidades cuando el precio sea bajo.

4. **Mitigación de la Volatilidad:** Al invertir de manera constante, DCA **mitiga el impacto de la volatilidad del mercado**. Los movimientos a corto plazo importan menos, ya que estás en el mercado de manera continua.

5. **Inversión Disciplinada:** DCA fomenta la **inversión disciplinada**, ya que te comprometes a invertir en intervalos regulares, sin importar cómo fluctúen los precios.

Si bien el DCA no garantiza ganancias, puede ser una **estrategia efectiva para quienes buscan invertir en criptomonedas sin intentar predecir los movimientos del mercado**. Al promediar los costos a lo largo del tiempo, es posible reducir el impacto de la volatilidad y posiblemente obtener un retorno positivo a largo plazo.

Es crucial **mantener la disciplina y seguir con las compras programadas** independientemente de las fluctuaciones del mercado a corto plazo. El DCA es una estrategia que se basa en la consistencia y en la confianza en el potencial a largo plazo de las criptomonedas.

**Beneficios:**

1. **Reducción del riesgo:** DCA ayuda a reducir la exposición al riesgo al distribuir las compras a lo largo del tiempo en lugar de realizar una gran inversión en un momento específico.
2. **Mitigación de la volatilidad:** Al comprar activos a intervalos regulares, se suavizan las fluctuaciones de precios a corto plazo, lo que puede ayudar a evitar la compra en momentos de alta volatilidad.
3. **Automatización:** DCA se puede automatizar, lo que permite a los inversores mantener una disciplina constante sin depender de decisiones emocionales en el mercado.
4. **Potencial para comprar a precios más bajos:** La estrategia permite comprar más unidades de un activo cuando los precios son bajos y menos unidades cuando son altos, lo que puede aumentar el promedio de compra a lo largo del tiempo.
5. **Aprovechamiento de la inversión en el tiempo:** DCA capitaliza el efecto del interés compuesto y el crecimiento a largo plazo, ya que el inversionista está adquiriendo activos a diferentes precios a lo largo del tiempo.

**Desventajas:**
1. **Oportunidades perdidas:** DCA podría llevar a perder oportunidades de comprar en momentos de baja y vender en momentos de alta si el mercado muestra una tendencia alcista constante.
2. **Altos costos de transacción:** Si se utilizan pequeñas inversiones en cada período, los costos de transacción pueden afectar el rendimiento neto.
3. **No es infalible:** Aunque reduce el riesgo, DCA no garantiza evitar pérdidas en un mercado en declive.
4. **Potencial para comprar en declives prolongados:** En un mercado bajista prolongado, la estrategia podría resultar en la adquisición de activos a precios bajos constantes, lo que podría generar pérdidas a largo plazo.
5. **Requiere paciencia:** DCA es una estrategia a largo plazo y requiere paciencia para ver resultados significativos en el rendimiento de la inversión.

## **1. Dominio del Mercado (%)**

**Introducción:**
El mundo de las criptomonedas ha brindado a los inversores una nueva dimensión en la diversificación de carteras. Sin embargo, en un mercado tan volátil y diverso, es esencial contar con métricas efectivas para evaluar la influencia de una criptomoneda específica en comparación con el mercado en general. Un indicador clave para medir esta influencia es el "Dominio del Mercado", un KPI (Indicador Clave de Desempeño) crucial para tomar decisiones informadas y establecer objetivos claros en el espacio de las inversiones en criptomonedas.

**Dominio del Mercado: Definición y Cálculo:**
El Dominio del Mercado de una criptomoneda es el porcentaje de la capitalización de mercado total que representa esa criptomoneda en relación con el mercado de todas las criptomonedas. El cálculo de este indicador se basa en dos datos fundamentales:

1. **Capitalización de Mercado Total:** Este valor es la suma de las capitalizaciones de mercado de todas las criptomonedas en circulación en un momento determinado. Se obtiene a partir de fuentes de datos en tiempo real, como CoinMarketCap, CoinGecko y otras plataformas similares.

2. **Capitalización de Mercado de la Criptomoneda Específica:** Este dato es la capitalización de mercado individual de la criptomoneda que se está evaluando.

El Dominio del Mercado se calcula mediante la fórmula:

**Dominio del Mercado (%) = (Capitalización de Mercado de la Criptomoneda Específica / Capitalización de Mercado Total) * 100**

Un alto Dominio del Mercado sugiere que la criptomoneda tiene una mayor influencia en el mercado, lo que podría indicar una posición más sólida en términos de adopción y valor relativo.

**Objetivo Ponderado y Medición:**
Establecer objetivos claros en función del Dominio del Mercado puede ser una estrategia efectiva para guiar tus decisiones de inversión. Uno de los enfoques sugeridos es el "Objetivo Ponderado", donde se establece un objetivo específico en términos de capitalización de mercado. Por ejemplo, puedes proponerte que las tres criptomonedas con la mayor capitalización de mercado en tu cartera representen al menos el 80% del Dominio Total.

En términos numéricos, esto se traduce en tres posibles niveles de cumplimiento:

1. **0% de Cumplimiento (50% de Capitalización):** Si tus tres principales criptomonedas no alcanzan al menos el 50% del Dominio Total, tu cumplimiento sería del 0%. Esto indicaría la necesidad de reequilibrar tu cartera para aumentar la influencia de estas criptomonedas.

2. **100% de Cumplimiento (70% de Capitalización):** Si logras que tus tres principales criptomonedas representen exactamente el 70% del Dominio Total, habrías alcanzado el 100% de cumplimiento. Esto significa que has logrado el equilibrio deseado según tu estrategia.

3. **150% de Cumplimiento (80% de Capitalización):** Si tus tres principales criptomonedas superan el objetivo y alcanzan el 80% del Dominio Total, tu cumplimiento sería del 150%. Esto podría indicar una concentración aún mayor en tus criptomonedas líderes.

**Conclusión:**
El Dominio del Mercado es un KPI crucial para evaluar la influencia de una criptomoneda en el mercado. Mediante el establecimiento de un Objetivo Ponderado, puedes mantener un equilibrio deseado entre criptomonedas influyentes y más pequeñas en tu cartera. Estos objetivos numéricos te brindan una base concreta para evaluar y ajustar tus inversiones, permitiéndote tomar decisiones informadas y estratégicas en un mercado en constante evolución. Recuerda que la flexibilidad y el monitoreo constante son esenciales para adaptarse a los cambios en el mercado de criptomonedas.

Se calcula el nivel de cumplimiento del KPI "dominio del mercado (%)" para un portafolio de inversión. Se suman los primeros 3 valores del dominio del mercado obtenidos del DataFrame y se compara con los valores objetivo para determinar el nivel de cumplimiento.

```python
suma_dominio_mercado = round(dominio_del_mercado['dominio_mercado'].sum(), 2)

# Definir los valores objetivo de dominio de mercado para cada escenario
target_dominio_mercado_0 = 50
target_dominio_mercado_100 = 70
target_dominio_mercado_150 = 80

# Calcular el nivel de cumplimiento basado en los valores objetivo de dominio de mercado
if suma_dominio_mercado >= target_dominio_mercado_100 and suma_dominio_mercado <= target_dominio_mercado_150:
    compliance_level = 100 + ((suma_dominio_mercado - target_dominio_mercado_100) / (target_dominio_mercado_150 - target_dominio_mercado_100)) * 50
elif suma_dominio_mercado > target_dominio_mercado_150:
    compliance_level = 150.0
else:
    compliance_level = ((suma_dominio_mercado - target_dominio_mercado_0) / (target_dominio_mercado_100 - target_dominio_mercado_0)) * 100

# Mostrar el resultado
print("El dominio del mercado (%) para el portafolio de inversión es:", suma_dominio_mercado)
print(f"Nivel de Cumplimiento: {compliance_level:.2f}%")
```

## **2. Índice de Correlación:**

El **Índice de Correlación** es un indicador clave en el mundo de las criptomonedas que revela cómo se comportan diferentes activos digitales en relación entre sí. Este KPI es fundamental para evaluar la interconexión y la dependencia entre las criptomonedas en tu cartera. Aquí está el proceso para medir este KPI:

**1. Recopilación de Datos de Precios:** Obtén datos históricos de precios de las criptomonedas que deseas comparar. Estos datos se pueden adquirir de diversas fuentes en línea, como plataformas de análisis de criptomonedas, exchanges o servicios de datos financieros.

**2. Cálculo de Rendimientos Diarios o Periódicos:** Utiliza los datos de precios para calcular los rendimientos diarios (o de otro período específico) para cada una de las criptomonedas. Los rendimientos se calculan como el cambio porcentual en el precio entre dos días consecutivos: **[(Precio Día 2 - Precio Día 1) / Precio Día 1] * 100.**

**3. Cálculo de la Correlación:** El paso más crucial implica calcular el índice de correlación en función de los rendimientos calculados anteriormente. Utiliza métodos estadísticos, como la correlación de Pearson, para obtener el valor de correlación. Este valor oscilará entre -1 y 1.

**4. Interpretación del Resultado:** El resultado de la correlación te proporcionará información sobre la relación entre las criptomonedas en tu cartera. Un valor cercano a 1 indica una correlación positiva fuerte, lo que significa que los precios de las criptomonedas tienden a moverse en la misma dirección. Un valor cercano a -1 indica una correlación negativa fuerte, lo que sugiere que los precios tienden a moverse en direcciones opuestas. Un valor cercano a 0 indica una correlación débil o nula, lo que significa que los precios no están relacionados.

**Niveles de Cumplimiento:**

2. **100% de Cumplimiento (Diversificación Equilibrada):** Si logras que la media de las correlaciones entre las tres criptomonedas de mayor capitalización sean todas 0.55, habrías alcanzado el 100% de cumplimiento. Esto significa que tu cartera está diversificada de manera equilibrada.

Esto indicaría una diversificación extrema, ya que las criptomonedas están completamente inversamente correlacionadas, lo que puede reducir el riesgo en ciertas condiciones.

## **3. Retorno de la Inversión (ROI)**

El **Retorno de la Inversión (ROI)** en el contexto de las criptomonedas es un indicador esencial que mide la rentabilidad de una inversión en términos de porcentaje. Los inversores utilizan el ROI para evaluar el rendimiento de sus inversiones y entender cuánto han ganado o perdido en relación con la cantidad de dinero que inicialmente invirtieron. Este indicador tiene un valor particular al tomar decisiones informadas sobre futuras inversiones y al comparar diversos activos digitales en función de su rentabilidad.

**Fórmula para Calcular el ROI:**

La fórmula para calcular el ROI en el contexto de las criptomonedas es la siguiente:

**ROI = (Ganancia - Costo) / Costo * 100**


**ROI = ((Valor Cripto - Valor de Inversión) / Valor de Inversión) * 100**

Donde:

- **Valor Cripto:** Es el valor actual de la criptomoneda después del período de inversión.
- **Valor de Inversión:** Es la cantidad de dinero invertida en la criptomoneda al inicio de la inversión.

**Cálculo del ROI a lo largo de 2 Años:**

En el marco de un vencimiento de inversión de 2 años, el cálculo del ROI adopta una aproximación específica. Se realiza tomando el promedio del ROI del primer año y el promedio del ROI del segundo año. En otras palabras, se calcula el promedio de los valores de ROI2 durante ambos años para obtener un único valor que representa la rentabilidad acumulada durante el período de 2 años.

**Interpretación del Resultado:**

El resultado del cálculo del ROI se expresa en forma de porcentaje. Un ROI positivo señala que la inversión ha generado ganancias, mientras que un ROI negativo indica una pérdida en la inversión. Un ROI del 0% implica que la inversión no ha experimentado cambio en su valor. Cuanto mayor sea el ROI positivo, mayor será la rentabilidad de la inversión en comparación con la cantidad inicial invertida.

**Importancia del ROI en la Toma de Decisiones:**

El ROI desempeña un papel crucial en la evaluación de la eficacia de las inversiones y en la comparación de diferentes oportunidades. Los inversores emplean el ROI para identificar qué criptomonedas o estrategias de inversión han sido más exitosas en el transcurso del tiempo. Además, el ROI brinda a los inversores la capacidad de ajustar y optimizar sus decisiones de inversión en función de los resultados obtenidos.

En resumen, el ROI es un indicador fundamental que suministra información valiosa sobre la rentabilidad de las inversiones en criptomonedas. Su habilidad para evaluar la eficiencia y comparar el rendimiento lo convierte en una herramienta esencial en el arsenal de cualquier inversor informado en el emocionante mercado de las criptomonedas.

Cargar el archivo CSV con la data histórica de precios de las criptomonedas con las que vamos a trabajar en un DataFrame.

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Cargar el archivo CSV en un DataFrame
data_price = pd.read_csv('Dataset//data_price.csv')

# Convertir la columna "Fecha" a formato DateTime y establecerla como índice
data_price["Fecha"] = pd.to_datetime(data_price["Fecha"])
data_price.set_index("Fecha", inplace=True)
```

Este código realiza un análisis detallado de diferentes criptomonedas, calculando y evaluando su desempeño financiero a lo largo del tiempo. A través de un proceso iterativo para cada criptomoneda, se determina cuántas unidades podrían haberse comprado con una inversión de $1000 USD en el último día de cada mes. Luego, se elabora un resumen anual que abarca aspectos como el valor total de inversión, unidades compradas, valor de la criptomoneda al cierre del año, y se calcula el retorno de la inversión (ROI).

Además, se aplican ajustes y refinamientos a los datos para mejorar la precisión de los indicadores. Estos incluyen la agregación de compras acumuladas y la corrección de valores en escenarios con ROI negativo. Los resultados de cada criptomoneda se guardan en archivos CSV individuales, que almacenan de manera organizada toda la información analizada. El proceso se lleva a cabo en un enfoque estructurado y automatizado, proporcionando una visión completa y cuantitativa del rendimiento de las criptomonedas en un contexto financiero.

```python
# Lista de criptomonedas
cryptocurrencies = ["Bitcoin", "Ethereum", "Binance Coin", "Aave", "Solana", "Polkadot", "Chainlink", "Avalanche", "Algorand", "Polygon"]

# Recorrer la lista de criptomonedas y ejecutar el código para cada una
for cryptocurrency in cryptocurrencies:

    # Filtrar solo los registros que corresponden al último día de cada mes
    last_day_of_month = data_price.resample('M').last()

    # Calcular cuántas unidades de Bitcoin se podrían haber comprado con $1000 USD
    investment_amount = 1000  # Inversión mensual en USD
    last_day_of_month['Units Bought'] = investment_amount / last_day_of_month[cryptocurrency]

    # Crear un DataFrame para mostrar los resultados
    investment_df = pd.DataFrame({
        'Fecha de Compra': last_day_of_month.index,
        'Precio': last_day_of_month[cryptocurrency].round(2),
        'Valor de Inversión (USD)': investment_amount,
        'Unidades Compradas': last_day_of_month['Units Bought'].round(2)
    })

    # Agregar una columna para el año
    investment_df['Año'] = investment_df['Fecha de Compra'].dt.year

    # Agregar una columna para el valor del Bitcoin al cierre de los 12 meses
    investment_df['Precio al Cierre de 12 Meses'] = investment_df['Fecha de Compra'].apply(
        lambda x: data_price[data_price.index.year == x.year].iloc[-1][cryptocurrency].round(2)
    )

    # Calcular resumen por año
    yearly_summary = investment_df.groupby('Año').agg({
        'Valor de Inversión (USD)': 'sum',
        'Unidades Compradas': 'sum',
        'Precio al Cierre de 12 Meses': 'first'  # Tomar el valor del primer mes
    }).round(2)

    # Calcular el valor actual de las unidades de Bitcoin al final de cada año
    for year in yearly_summary.index:
        year_end_price = data_price.loc[data_price.index.year == year, cryptocurrency].iloc[-1].round(2)
        yearly_summary.loc[year, 'Valor Cripto'] = (year_end_price * yearly_summary.loc[year, 'Unidades Compradas']).round(2)

    # Calcular el ROI al final de cada año
    yearly_summary['ROI'] = ((yearly_summary['Valor Cripto'] - yearly_summary['Valor de Inversión (USD)']) / yearly_summary['Valor de Inversión (USD)']) * 100
    yearly_summary['ROI'] = yearly_summary['ROI'].round(2)  # Redondear a 2 decimales

    # Agregar la columna ROI2
    yearly_summary['ROI2'] = yearly_summary.apply(lambda row: 0 if row['Valor Cripto'] < row['Valor de Inversión (USD)'] else row['ROI'], axis=1)

    # Agregar la columna "Unidades Compradas2" con los mismos valores que "Unidades Compradas"
    yearly_summary["Unidades Compradas2"] = yearly_summary["Unidades Compradas"]

    for i in range(1, len(yearly_summary)):
        if yearly_summary.at[yearly_summary.index[i - 1], "ROI2"] == 0:
            yearly_summary.at[yearly_summary.index[i], "Unidades Compradas2"] += yearly_summary.at[yearly_summary.index[i - 1], "Unidades Compradas2"]

    # Actualizar "Unidades Compradas2" a 0 cuando "ROI2" es 0
    yearly_summary.loc[yearly_summary['ROI2'] == 0, 'Unidades Compradas2'] = 0

    # Agregar la columna "Valor Cripto2"
    yearly_summary['Valor Cripto2'] = (yearly_summary['Unidades Compradas2'] * yearly_summary['Precio al Cierre de 12 Meses']).round(2)


    # Agregar la columna "Valor de Inversión (USD)2" con los mismos valores que "Valor de Inversión (USD)"
    yearly_summary["Valor de Inversión (USD)2"] = yearly_summary["Valor de Inversión (USD)"]

    for i in range(1, len(yearly_summary)):
        if yearly_summary.at[yearly_summary.index[i - 1], "ROI2"] == 0:
            yearly_summary.at[yearly_summary.index[i], "Valor de Inversión (USD)2"] += yearly_summary.at[yearly_summary.index[i - 1], "Valor de Inversión (USD)2"]

    # Actualizar "Valor de Inversión (USD)2" a 0 cuando "ROI2" es 0
    yearly_summary.loc[yearly_summary['ROI2'] == 0, 'Valor de Inversión (USD)2'] = 0

    # Calcular el ROI2 utilizando las columnas "Valor Cripto2" y "Valor de Inversión (USD)2"
    yearly_summary['ROI2'] = ((yearly_summary['Valor Cripto2'] - yearly_summary['Valor de Inversión (USD)2']) / yearly_summary['Valor de Inversión (USD)2']) * 100
    yearly_summary['ROI2'] = yearly_summary['ROI2'].round(2)  # Redondear a 2 decimales

    # Reemplazar los valores NaN en la columna "ROI2" con 0
    yearly_summary['ROI2'] = yearly_summary['ROI2'].fillna(0)


    # Calcular el ROI en USD y agregar la columna correspondiente
    yearly_summary['ROI en USD'] = (yearly_summary['Valor Cripto'] - yearly_summary['Valor de Inversión (USD)']).round(2)

    # Calcular el ROI en USD y agregar la columna correspondiente
    yearly_summary['ROI en USD2'] = (yearly_summary['Valor Cripto2'] - yearly_summary['Valor de Inversión (USD)2']).round(2)

    # Definir el nuevo orden de columnas
    new_column_order = [
        'Precio al Cierre de 12 Meses',
        'Valor de Inversión (USD)', 'Unidades Compradas',
        'Valor Cripto', 'ROI', 'ROI en USD',
        'Valor de Inversión (USD)2', 'Unidades Compradas2',
        'Valor Cripto2', 'ROI2', 'ROI en USD2'
    ]

    # Reorganizar las columnas del DataFrame
    yearly_summary = yearly_summary[new_column_order]

    # Insertar columna vacía como separador
    yearly_summary.insert(yearly_summary.columns.get_loc('Valor de Inversión (USD)'), '-------', '')

    # Insertar columna vacía como separador
    yearly_summary.insert(yearly_summary.columns.get_loc('Valor de Inversión (USD)2'), '------', '')

    # Eliminar filas con valores NaN
    yearly_summary = yearly_summary.dropna()

    # Filtrar solo los registros con ROI en USD2 menor que 0
    negative_roi_rows = yearly_summary[yearly_summary['ROI en USD2'] < 0]

    # Actualizar las columnas en las filas con ROI en USD2 menor que 0
    yearly_summary.loc[negative_roi_rows.index, ['Valor de Inversión (USD)2', 'Unidades Compradas2', 'Valor Cripto2', 'ROI2', 'ROI en USD2']] = 0

    # Crear una nueva columna con el nombre de la criptomoneda
    yearly_summary["Cryptocurrency"] = cryptocurrency

    # Ruta base donde se guardarán los archivos CSV
    base_path = "Dataset/DF_ROI/"

    # Nombre del archivo basado en la criptomoneda seleccionada
    csv_filename = f"ROI_{cryptocurrency}.csv"

    # Ruta completa del archivo CSV
    csv_path = base_path + csv_filename

    # Guardar el DataFrame en un archivo CSV
    yearly_summary.to_csv(csv_path, index=True)

    # Imprimir mensaje de progreso
    print(f"Archivo CSV para {cryptocurrency} creado exitosamente.")

print("Proceso completado.")
```

Se realiza la importación y consolidación de varios DataFrames que contienen información de retorno de inversión (ROI) de diferentes criptomonedas para una visión más completa y conveniente de la información.

- **Precio al Cierre de 12 Meses:** El precio de la criptomoneda al final del último día del mes en el que se realizó la inversión.

- **Valor de Inversión (USD):** El valor total invertido en la criptomoneda durante el año correspondiente.

- **Unidades Compradas:** La cantidad total de unidades de la criptomoneda compradas durante el año correspondiente.

- **Valor Cripto:** El valor total actual de las unidades de la criptomoneda al final del año, basado en el precio al cierre de 12 meses y las unidades compradas.

- **ROI en USD:** El retorno de la inversión en dólares, calculado como Valor Cripto - Valor de Inversión.

- **Valor de Inversión (USD)2:** El valor total invertido acumulado en la criptomoneda hasta el final del año, considerando los años anteriores con ROI negativo.

- **Unidades Compradas2:** La cantidad total de unidades acumuladas de la criptomoneda hasta el final del año, considerando los años anteriores con ROI negativo.

- **Valor Cripto2:** El valor total actual acumulado de las unidades de la criptomoneda al final del año, considerando los años anteriores con ROI negativo.

- **ROI2:** El retorno de la inversión acumulado en porcentaje, calculado utilizando los valores de Valor Cripto2 y Valor de Inversión (USD)2.

- **ROI en USD2:** El retorno de la inversión acumulado en dólares, calculado como Valor Cripto2 - Valor de Inversión (USD)2.
